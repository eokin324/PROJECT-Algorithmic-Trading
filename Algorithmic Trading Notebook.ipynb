```python
df=pd.read_excel('S&P500.xlsx')
list_of_stocks=list(df['Symbol'])
list_of_stocks
start=datetime.datetime(2015,1,1) 
end=datetime.datetime(2017,12,30)
stocks = web.get_data_yahoo(list_of_stocks,start,end,interval='m')['Adj Close']
stocks
```

```python
stocks.dropna(axis=1, how='all',inplace=True)
```

```python
stocks.head()
```

```python
print('Number of companies to analyze: ', len(stocks.columns))
```

```python
stocks.pct_change(1).mean().sort_values(ascending=False).drop(['INFO','WELL','DXC','LW']).head(10)
```

```python
ten_stocks=list(stocks.pct_change(1).mean().sort_values(ascending=False).drop(['INFO','WELL','DXC','LW']).head(10).index)
```

```python
stocks=stocks[ten_stocks]
```

```python
stocks
```

```python
log_returns = np.log(stocks/stocks.shift(1))
log_returns.head(3)
```

```python
np.random.seed(88)
print(stocks.columns)
weights=np.array(np.random.random(len(stocks.columns)))
print('Random Weights:')
print(weights)
print('Rebalanced Weights:')
weights = weights/np.sum(weights) #normalize
print(weights)
```

```python
print('Expected Portfolio Return')
expected_return = np.sum(log_returns.mean() * weights * 12)
print(expected_return)
```

```python
print('Expected Volatility')
expected_volatility = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*12,weights)))
print(expected_volatility)
```

```python
print('Sharpe Ratio')
SR = expected_return / expected_volatility
print(SR)
```

```python
number_portfolios = 5000
all_weights = np.zeros((number_portfolios,len(stocks.columns)))
returns_array = np.zeros(number_portfolios)
volatility_array = np.zeros(number_portfolios)
sharpe_array = np.zeros(number_portfolios)

for i in range(number_portfolios):
    weights=np.array(np.random.random(len(stocks.columns)))
    weights=weights/np.sum(weights)
    all_weights[i,:] = weights #save weights
    returns_array[i] = np.sum ((log_returns.mean() * weights) * 12)
    volatility_array[i] = np.sqrt(np.dot(weights.T,np.dot(log_returns.cov()*12,weights)))
    sharpe_array[i] = returns_array[i]/volatility_array[i]
```

```python
sharpe_array.max()
```

```python
sharpe_array.argmax()
```

```python
optimal_weights=all_weights[sharpe_array.argmax(),:]
```

```python
df=pd.DataFrame(optimal_weights,index=stocks.columns)
df.rename(columns = {0:'Optimal Weights'},inplace = True)
df=df[df['Optimal Weights']!=0]
df
```
